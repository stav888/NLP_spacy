```python
# Task 1
import spacy
nlp = spacy.load("en_core_web_sm")

doc8 = nlp("Serena Williams had dinner with Tom Hanks in Paris.")

for entity in doc8.ents:
    print(f'{entity.text:<17} {entity.label_}', '\n')
```

```python
# Task 2

import spacy
nlp = spacy.load("en_core_web_sm")

text = nlp("Serena Williams had dinner with Tom Hanks in Paris.")

for entity in text.ents:
    if entity.label_ == "PERSON":
        print(entity.text)
```

```python
# Task 3
import spacy
nlp = spacy.load("en_core_web_sm")

text = ("She was running and had run 5 kilometers by 7am.")

doc = nlp(text)
for token in doc:
    print(f'{token.text:<10} {token.lemma_}')
```

```python
# Task 4

import spacy
from spacy.lang.en.stop_words import STOP_WORDS

nlp = spacy.load("en_core_web_sm")
text = "This is an example sentence with some stop words."
doc = nlp(text)

for token in doc:
    #print(token.text, token.text.lower() in STOP_WORDS)
    if token.text.lower() not in STOP_WORDS:
        print(f'{token.text}')
```

```python
# Task 5

import spacy
from spacy.lang.en.stop_words import STOP_WORDS

nlp = spacy.load("en_core_web_sm")
text = "SpaCy is awesome and powerful."
doc = nlp(text)

for token in doc:
    if token.text.lower() in STOP_WORDS:
        print(f'{token.text}')

nlp.Defaults.stop_words.add('powerful')
nlp.vocab['powerful'].is_stop = True
```

```python
# Task 6
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)

pattern = [nlp("Artificial Intelligence"), nlp("artificial intelligence"), nlp("artificial-intelligence")]
matcher.add("ARTIFICIAL_INTELLIGENCE", pattern)

docs = nlp("Artificial Intelligence is the future. I study artificial intelligence.")
matches = matcher(docs)

for match_id, start, end in matches:
    print(docs[start:end].text)
```

```python
# Task 7
import spacy

nlp = spacy.load("en_core_web_sm")

text = nlp("The cat sat on the mat.")
doc = nlp(text)

for token in doc:
    print(f"{token.text:<10} {token.pos_:<10} {spacy.explain(token.pos_)}")
```

```python
# Task 8
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

user = input("enter a sentence: ")
doc = nlp(user)

for token in doc:
    print(f"{token.text:<10} {token.pos_:<10} {spacy.explain(token.pos_)}")
    
displacy.render(doc, style="dep")

#Apple is looking at buying a U.K. startup for $1 billion.
```
